```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Colored Scatterplots

```python
iris = pd.read_csv('../DataSets/iris.csv')
iris.sample(5)
```

```python
iris['Species'] = iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
iris.sample(5)
```

- we can use the `c` (color) parameter to color the points based on a third variable  , and we can use the `cmap` parameter to specify a colormap.
- we can also use the `alpha` parameter to adjust the transparency of the points.
- colormaps can be found [here](https://matplotlib.org/stable/tutorials/colors/colormaps.html)
- we can use `plt.colorbar()` to add a colorbar to the plot, which shows the mapping of colors to values.

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='jet',alpha=0.7)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

### Plot size

```python
plt.figure(figsize=(15,7))

plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='jet',alpha=0.7)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

### Subplots

```python
# A different way to plot graphs
batters.head()
```

```python
# previous way
plt.figure(figsize=(22,8))
plt.scatter(batters['avg'],batters['strike_rate'])
plt.title('Something')
plt.xlabel('Avg')
plt.ylabel('Strike Rate')

plt.show()
```

```python
batters
```

- `fig` : the Figure object (overall canvas that can hold multiple subplots)
- `ax` : an Axes (individual plot area). you draw on ax (plot, set title/labels).
- Use fig to adjust overall layout or save the image, and ax to control the specific plot.

```python
# another way
fig,ax = plt.subplots(figsize=(22,8))

ax.scatter(batters['avg'],batters['strike_rate'],color='red',marker='o',s=batters['runs']/7,alpha=0.7,edgecolor='black',linewidth=1)
ax.set_title('Something')
ax.set_xlabel('Avg')
ax.set_ylabel('Strike Rate')

fig.show()
```

```python
fig, ax = plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(10,6))

ax[0].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[1].scatter(batters['avg'],batters['runs'])

ax[0].set_title('Avg Vs Strike Rate')
ax[0].set_ylabel('Strike Rate')

ax[1].set_title('Avg Vs Runs')
ax[1].set_ylabel('Runs')
ax[1].set_xlabel('Avg')
```

```python
# now ax is a 2d array so we can access individual plots using indexing like ax[0,0], ax[0,1] etc
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,10))

ax[0,0].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[0,1].scatter(batters['avg'],batters['runs'])
ax[1,0].hist(batters['avg'])
ax[1,1].hist(batters['runs'])
```

```python
# alternative way to do the same thing  (but above way is preferred)

fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)    # first 2 is rows and columns and last 1 is the index of the plot
ax1.scatter(batters['avg'],batters['strike_rate'],color='red')

ax2 = fig.add_subplot(2,2,2)    # first 2 is rows and columns and last 2 is the index of the plot
ax2.hist(batters['runs'])

ax3 = fig.add_subplot(2,2,3)    # first 2 is rows and columns and last 3 is the index of the plot
ax3.hist(batters['avg'])
```

```python
fig, ax = plt.subplots(nrows=2,ncols=2,sharex=True,figsize=(10,10))
# here ax is a 2d array so we can access individual plots using indexing like ax[0,0], ax[0,1
ax[1,1] = plt.subplot(2,2,3)
ax[1,1].scatter(batters['strike_rate'],batters['runs'],color='green' , s  = batters['runs']/10 , alpha=0.5, edgecolor='black', linewidth=1)
ax[1,1].set_title('Strike Rate Vs Runs')
ax[1,1].set_xlabel('Strike Rate')
ax[1,1].set_ylabel('Runs')
```

### 3D Surface Plots

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx, yy = np.meshgrid(x,y)
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.log(xx)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

### Pandas Plot()

- Pandas has a built-in plotting function that uses Matplotlib under the hood.
- You can call the `plot()` method on a DataFrame or Series to create various types of plots.
- You can specify the type of plot using the `kind` parameter (e.g., 'line', 'bar', 'hist', 'box', 'scatter', etc.).
- You can customize the plot using additional parameters like `title`, `xlabel`, `ylabel`, `color`, etc.
- Pandas plotting is convenient for quick visualizations, but for more complex plots, using Matplotlib directly is recommended.

```python
# on a series

s = pd.Series([1,2,3,4,5,6,7])
s.plot(kind='pie')
```
