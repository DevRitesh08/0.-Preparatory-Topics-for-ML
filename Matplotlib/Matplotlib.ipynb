```python
# from a pandas dataframe
batsman = pd.read_csv('../DataSets/sharma-kohli.csv')
batsman

plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
# plotting multiple plots
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# labels title
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# colors(hex) and line(width and style) and marker(size)
plt.plot(batsman['index'],batsman['V Kohli'],color='red') # can also use hex code in color #D9F10F , #FC00D6
plt.plot(batsman['index'],batsman['RG Sharma'],color='purple')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3)     # linestyle parameter can take values like dasdot , solid and dashed .
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2)

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2,marker='o')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# legend -> location
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3,marker='D',markersize=10,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2,marker='o',label='Rohit')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.legend(loc = 'best')       # can also change the position of legend using the parameter loc , by default it is set to best ==> it auto. finds the best location .
# Supported values of loc are 'best', 'upper right', 'upper left', 'lower left', 'lower right', 'right', 'center left', 'center right', 'lower center', 'upper center', 'center
```

```python
import matplotlib.pyplot as plt

price = [48000, 54000, 57000, 49000, 47000, 45000, 4500000]
year = [2015, 2016, 2017, 2018, 2019, 2020, 2021]

plt.plot(year, price)

# The price data contains a significant outlier (4,500,000) that makes the
# variation in other data points difficult to see. By limiting the y-axis,
# we can focus on the majority of the data for a more readable plot.
plt.ylim(0, 75000)

# We also limit the x-axis to zoom in on a specific period of interest.
plt.xlim(2017, 2019)

# To display the plot
plt.show()
```

# Scatter Plots

image.png

- Bivariate Analysis    : means to study two columns at the same time .
- Best for numerical vs numerical data .
- Use case - Finding correlation
- 2d plot is the special case of scatter plot , since their we also join the points with lines .

```python
# plt.scatter simple function
x = np.linspace(-10,10,50)

y = 10*x + 3 + np.random.randint(0,300,50)      # y = mx + c + noise(randomit)
y
```

```python
plt.scatter(x,y)
```

```python
# plt.scatter on pandas data
df = pd.read_csv('../DataSets/batter.csv')
df = df.head(50)
df
```

```python
plt.scatter(df['avg'],df['strike_rate'],color='red',marker='*',s=100)   # s is for size of marker
plt.title('Avg and SR analysis of Top 50 Batsman')
plt.xlabel('Average')
plt.ylabel('SR')
```

```python
# size
tips = sns.load_dataset('tips')


# slower
plt.scatter(tips['total_bill'],tips['tip'],s=tips['size']*20)   # here parameter s takes an array like input which defines the size of each point in the scatter plot , which is scaled by 20 here .
# this scatter plot shows the relation between total bill and tip amount , and the size of the point shows the size of the table or number of people in the table , hence it shows relation between these three variables .
```

```python
# scatterplot using plt.plot
# faster , but less features like size and color mapping and all .
plt.plot(tips['total_bill'],tips['tip'],'*')
```

### Bar chart

image.png

- Bivariate Analysis

```python
# simple bar chart
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']

plt.bar(colors,children,color='purple')
```

```python
# bar chart using data
```

```python
# color and label
df = pd.read_csv('../DataSets/batsman_season_record.csv')
df
```

### Multiple Bar charts

```python
plt.bar(df['batsman'],df['2015'],color='magenta')
plt.bar(df['batsman'],df['2016'],color='orange')
plt.bar(df['batsman'],df['2017'],color='green')
# hence it shows the runs scored by each batsman in each season , but it is difficult to compare the runs scored by each batsman in different seasons as the bars are stacked on top of each other . To overcome this we can use the parameter bottom to stack the bars on top of each other .
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2,df['2015'],width=0.2,color='green')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='orange')
plt.bar(np.arange(df.shape[0]) + 0.2,df['2017'],width=0.2,color='pink')

plt.xticks(np.arange(df.shape[0]), df['batsman'])

plt.show
```

```python
np.arange(df.shape[0])
```

```python
# a problem
children = [10,20,40,10,30]
colors = ['red red red red red red','blue blue blue blue','green green green green green','yellow yellow yellow yellow ','pink pinkpinkpink']

plt.bar(colors,children,color='black')
# plt.xticks(rotation='vertical')
```

```python
# Stacked Bar chart
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016'] + df['2017']),label='2015')
# plt.grid()

plt.legend()
plt.show()
```

### Pie Chart

image.png

- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

```python
data = [23,45,100,20,49,50,60,21]
plt.pie(data)
plt.show
```

```python
# simple data
data = [23,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data,labels=subjects)

plt.show()
```

```python
# dataset
df = pd.read_csv('../Datasets/gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'])
plt.show()
```

```python
# percentage and colors
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['blue','green','yellow','pink','cyan','brown'])
plt.show()
```

```python
# explode shadow
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.1,0,0.0,0,0,0.2],shadow=True)
plt.show()
```
