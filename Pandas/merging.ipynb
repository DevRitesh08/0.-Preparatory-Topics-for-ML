## JOINS

- Joins are used to combine two dataframes based on common columns or indices.
- SQL style joins are implemented in pandas using the merge function.
    - merge function : it takes 3 main parameters
        - how : type of join
        - on : column name on which join is to be performed
        - left_on , right_on : if the column names are different in both the dataframes then we can use these parameters to specify the column names .
- types of joins
    - inner : returns only the rows with matching values in both dataframes .
    - left : returns all the rows from the left dataframe and the matching rows from the right dataframe . if there is no match then NaN values are filled in the right dataframe columns .
    - right : returns all the rows from the right dataframe and the matching rows from the left dataframe . if there is no match then NaN values are filled in the left dataframe columns .
    - outer : returns all the rows from both dataframes . if there is no match then NaN values are filled in the columns of the dataframe which does not have the match .

`inner join`
- by default merge function performs inner join .

```python
students.merge(regs,how='inner',on='student_id')
```

- inner join returns only the rows with matching values in both dataframes  , so the number of rows in the result will be less than or equal to the number of rows in both dataframes .
- in the above example , only those students who have enrolled in any course are returned .

`left join`

```python
courses.merge(regs,how='left',on='course_id')
```

`# right join`

```python
temp_df = pd.DataFrame({
    'student_id':[26,27,28],
    'name':['Rohit','Shantanu','Rajesh'],
    'partner':[28,26,27]
})

students = pd.concat([students,temp_df],ignore_index=True)
```

```python
students.tail()
```

```python
students.merge(regs,how='right',on='student_id')
```

- so by interchanging the dataframes we can perform right join using left join and vice versa .

```python
regs.merge(students,how='left',on='student_id')
```

`outer join`

```python
students.merge(regs,how='outer',on='student_id').tail(10)
```

Q : find month by month revenue

```python
temp_df = pd.concat([nov,dec],keys=['Nov','Dec']).reset_index()
temp_df.merge(courses,on='course_id').groupby('level_0')['price'].sum()
# level_0 is the column created by reset_index() which contains the keys used in concat function .
```

Q : Print student name -> partner name for all enrolled students

```python
# self join : joining a dataframe with itself is called self join .
# here we are joining students dataframe with itself to get the partner name for each student .
students.merge(students,how='inner',left_on='partner',right_on='student_id')[['name_x','name_y']]
```

Q : find top 3 students who did most number enrollments


```python
regs.merge(students,on='student_id').groupby(['student_id','name'])['name'].count().sort_values(ascending=False).head(3)
# here we have grouped by both student_id and name to avoid same name students getting merged together .
```

Q : find top 3 students who spent most amount of money on courses


```python
regs.merge(students,on='student_id').merge(courses,on='course_id').groupby(['student_id','name'])['price'].sum().sort_values(ascending=False).head(3)
```
