# **groupby**

- groupby() is a powerful function in pandas that allows you to split your data into groups based on some criteria, apply a function to each group, and then combine the results back into a single DataFrame. This is particularly useful for performing operations like aggregation, transformation, and filtering on subsets of your data.
- groups are formed on the data based on the values in one or more columns , generally we have two types of groups
    - **Categorical groups**: These are groups formed based on categorical variables, such as 'Genre', 'Country', or 'Director'. Each unique value in the categorical column defines a group.
    - **Numerical groups**: These are groups formed based on numerical variables, such as 'Year' or 'Rating'. You can create bins or ranges to categorize numerical data into groups.
- but groupby() does not work with the column having numeric data type .

```python
movies.groupby('Genre').ngroups
```

```python
movies['Genre'].nunique()
```

- find items in each group

```python
movies['Genre'].value_counts()
# here size() and value_counts() both give the same result but value_counts() is more efficient as it directly counts the occurrences of each unique value in the 'Genre' column without the overhead of creating a groupby object .
```

`first() , last() and nth()`
- These methods allow you to retrieve the first, last, or nth item from each group.
- They are useful for quickly accessing specific rows within each group without having to iterate through the groups manually.

```python
genres = movies.groupby('Genre')
```

```python
genres.first()
# This will return the first row of each group based on the order of the original DataFrame.
```

```python
genres.last()
# This will return the last row of each group based on the order of the original DataFrame
```

```python
genres.nth(2)
# This will return the third row (index 2) of each group. If a group has fewer than three rows, it will not appear in the result.
```

`get_group()`
- This method allows you to retrieve a specific group from the grouped object.
- You need to pass the group name (the value of the column you grouped by) as an argument to this method.

```python
genres.get_group('Western')
# This will return all rows where the 'Genre' is 'Western'.
```

```python
# another way
movies[movies['Genre'] == 'Western']
```

- Here, get_group() is more efficient as it directly accesses the specified group without filtering the entire DataFrame.

`groups`
- This attribute returns a dictionary where the keys are the group names and the values are the indices of the rows that belong to each group.

```python
genres.groups
```

`describe()`
- This method provides a summary of statistics for each group, similar to the describe() method for a DataFrame.

```python
genres.describe()
```

`sample()`
- This method allows you to randomly sample a specified number of rows from each group.

```python
genres.sample(2 , replace=True)
# This will return 2 random rows from each genre group.
# Samples 2 rows per group; raises ValueError if any group has <2 rows. Use replace=True or sample up to min(len(g), 2).
```

`nunique()`
- This method returns the number of unique values in each group for each column.

```python
genres.nunique()
```

## agg method
- The agg() method allows you to apply multiple aggregation functions to different columns in a single call.
- You can pass a dictionary where the keys are column names and the values are lists of functions to apply to those columns.

```python
genres.agg(
    {
        'Runtime':'mean',
        'IMDB_Rating':'mean',
        'No_of_Votes':'sum',
        'Gross':'sum',
        'Metascore':'min'
    }
)
```

```python
# passing list : apply multiple functions to all columns
genres.agg(['min','max','mean'])
```

```python
# apply \[`min`,`max`,`mean`\] only to numeric columns to avoid TypeError on object dtypes
numeric_cols = movies.select_dtypes(include='number').columns
movies.groupby('Genre')[numeric_cols].agg(['min', 'max', 'mean'])
```

```python
# Adding both the syntax
genres.agg(
    {
        'Runtime':['min','mean'],
        'IMDB_Rating':'mean',
        'No_of_Votes':['sum','max'],
        'Gross':'sum',
        'Metascore':'min'
    }
)
```

### Looping through groups :
- You can iterate over each group in the grouped object using a for loop.
- Each iteration returns a tuple containing the group name and the corresponding DataFrame for that group.

```python
for group,data in genres :
    print(f'Group: {group}')
    print(data.head(2))  # Display the first 2 rows of each group
    print('\n')
```

Q :  find the highest rated movie of each genre .

```python
frames = [
    data[data['IMDB_Rating'] == data['IMDB_Rating'].max()]
    for _, data in genres
]
df = pd.concat(frames, ignore_index=True)
df
# frames is a list of DataFrames, each containing the highest rated movie for a specific genre. pd.concat(frames, ignore_index=True) combines these individual DataFrames into a single DataFrame df, with the index reset to ensure a continuous sequence.
# ignore_index=True is used to reset the index in the concatenated DataFrame, providing a clean, sequential index starting from 0.
```

`# split (apply) combine strategy`
- The split-apply-combine strategy is a powerful data analysis technique used in pandas to efficiently process and analyze data by breaking it down into manageable parts.
- The strategy involves three main steps:
    1. **Split**: The data is divided into groups based on one or more categorical variables using the groupby() function. Each group contains all the rows that share the same values in the specified columns.
    2. **Apply**: A function is applied to each group independently. This function can be an aggregation (like sum, mean, etc.), a transformation (like scaling or normalizing), or a custom function defined by the user. The function processes the data within each group separately.
    3. **Combine**: The results from the applied function are then combined back into a single DataFrame or Series. This step consolidates the outputs from each group into a cohesive structure that can be easily analyzed or visualized.
- This strategy is particularly useful for handling large datasets, as it allows for efficient computation by leveraging the inherent structure of the data.

```python
genres.apply(min)
```

Q : find number of movies starting with A for each group

```python
def noofa(group):
  return group['Series_Title'].str.startswith('A').sum()
```

```python
genres.apply(noofa , include_groups=False)
# include_groups=False is used to exclude the group names from the index of the resulting Series, providing a cleaner output that focuses solely on the computed values.
```

Q : find ranking of each movie in the group according to IMDB score

```python
def rank_movie(group):
  group['genre_rank'] = group['IMDB_Rating'].rank(ascending=False)
  return group
```

```python
genres.apply(rank_movie)
```

Q : find normalized IMDB rating group wise

```python
def normal(group):
  group['norm_rating'] = (group['IMDB_Rating'] - group['IMDB_Rating'].min())/(group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
  return group

genres.apply(normal)
```

Q : Create a function that can return the highest score of any batsman


```python
temp_df = ipl[ipl['batsman'] == 'V Kohli']
temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
```

```python
def highest(batsman):
  temp_df = ipl[ipl['batsman'] == batsman]
  return temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
```
