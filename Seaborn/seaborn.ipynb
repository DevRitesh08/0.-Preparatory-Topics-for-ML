### Why Seaborn?

- provides a layer of abstraction over matplotlib , hence easy to use.
- better aesthetics
- more graphs included

### Seaborn Roadmap

- we can form a same graph using two different functions that is figure level and axis level function.

Types of Functions

- Figure Level : it creates a new figure and set of axes for each plot , it is a higher level function.
- Axis Level : it draws onto a specific matplotlib axes of an existing figure , it is a lower level function.

Main Classification

- Relational Plot
- Distribution Plot
- Categorical Plot
- Regression Plot
- Matrix Plot
- Multiplots

### 1. Relational Plot

- to see the statistical relation between 2 or more variables.
- Bivariate Analysis

Plots under this section

- scatterplot
- lineplot

- Figure Level Functions are recommmended for most cases.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
# scatter plot -> axes level function
# here we have to explicitly mention data = tips not like in plotly .
sns.scatterplot(data=tips, x='total_bill', y='tip',hue='sex',style='time',size='size')
# here we can show multiple information using hue , style and size parameter.
# here we can observe that the legend  , x axis and y axis title are automatically added.
```

```python
# relplot -> figure level -> by default in square shape
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter',hue='sex',style='time',size='size')
```

- hue : it will change the color of the points based on the categorical variable
- size : it will change the size of the points based on the numerical variable
- style : it will change the shape of the points based on the categorical variable

```python
# line plot
# it is used to see the trend of the data over a period of time
gap = px.data.gapminder()
gap
```

```python
ind = gap[gap['country'] == 'India']
ind
```

```python
# axes level function
sns.lineplot(data= ind , x='year', y='lifeExp')
```

```python
# using relplot
sns.relplot(data= ind , x='year', y='lifeExp', kind='line')
```

```python
# hue -> style
temp_df = gap[gap['country'].isin(['India','Brazil','Germany'])]
temp_df
```

```python
temp_df
```

```python
sns.relplot(kind='line', data=temp_df, x='year', y='lifeExp', hue='country', style='continent', size='continent')
```

- we can observe that in figure level function the legend is outside the graph , whereas in axes level function the legend is inside the graph.

```python
sns.lineplot(data=temp_df, x='year', y='lifeExp', hue='country')
```

```python
# facet plot -> figure level function -> work with relplot
# it will not work with scatterplot and lineplot since they are axes level function .
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', col='sex')
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', col='sex', row='day')
# same can be done for line plot
```

```python
# col wrap
sns.relplot(data=gap, x='lifeExp', y='gdpPercap', kind='scatter', col='year', col_wrap=3)
```

```python
# sns.scatterplot(data=tips, x='total_bill', y='tip', col='sex', row='day')
# this will give error since scatterplot is axes level function
```

#### Difference between facet plot and subplots
- Facet plot creates multiple panels that each show the same type of plot for different subsets of the data.
- Subplots create multiple panels that can each show different types of plots.

#### Summery
- Figure level function will create a new figure and set of axes for each plot.
- Axes level function will draw onto a specific matplotlib axes of an existing figure.
- Figure level function is a higher level function.
- Axes level function is a lower level function.
- Figure level function is recommended for most cases.
- Axes level function is more flexible.

### 2. Distribution Plots

- used for univariate analysis
- used to find out the distribution
- Range of the observation
- Central Tendency
- is the data bimodal?
- Are there outliers?

Plots under distribution plot

- histplot
- kdeplot
- rugplot

```python
# figure level function -> displot
# axes level function -> histplot -> kdeplot -> rugplot
```

```python
# plotting univariate histogram
sns.histplot(data=tips, x='total_bill')
```

```python
sns.displot(data=tips, x='total_bill', kind='hist')
```

```python
# bins parameter
sns.displot(data=tips, x='total_bill', kind='hist',bins=20)
```

```python
# Itâ€™s also possible to visualize the distribution of a categorical variable using the logic of a histogram.
# Discrete bins are automatically set for categorical variables

# here histogram will count the number of occurrences of each category so it can be called as count plot
sns.displot(data=tips, x='day', kind='hist')
```

```python
# hue parameter
sns.displot(data=tips, x='tip', kind='hist',hue='sex')
```

```python
# element -> step : it will draw the histogram as a step function so that we can see the distribution more clearly
sns.displot(data=tips, x='tip', kind='hist',hue='sex',element='step')
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.displot(data=titanic, x='age', kind='hist',element='step',hue='sex')
```

```python
# faceting using col and row -> not work on histplot function
sns.displot(data=tips, x='tip', kind='hist',col='sex',element='step' , color='m')
```

```python
# kdeplot (kernel density estimate plot)
# A KDE plot visualizes the distribution of observations in a dataset, similar to a histogram.
# Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate
sns.kdeplot(data=tips,x='total_bill')
```

```python
sns.displot(data=tips,x='total_bill',kind='kde')
```

```python
# hue -> fill
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex',fill=True,height=10,aspect=2)
```

```python
# Rugplot

# Plot marginal distributions by drawing ticks along the x and y axes.
# This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.
# unobtrusive means it will not cover the main plot

sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
```

```python
# Bivariate histogram
# A bivariate histogram bins the data within rectangles that tile the plot
# and then shows the count of observations within each rectangle with the fill color
# it is also called as 2D histogram where the intensity of color represents the count of observations in that bin .

# sns.histplot(data=tips, x='total_bill', y='tip')
sns.displot(data=tips, x='total_bill', y='tip',kind='hist')
```

```python
# Bivariate Kdeplot
# a bivariate KDE plot smoothes the (x, y) observations with a 2D Gaussian
# here the closer the contour lines are , the higher the density of points in that region .
sns.kdeplot(data=tips, x='total_bill', y='tip')
```

### 2. Matrix Plot

- Heatmap
- Clustermap

```python
# Heatmap
# Plot rectangular data as a color-encoded matrix
temp_df = gap.pivot(index='country',columns='year',values='lifeExp')

# axes level function , and their so no figure level function for heatmap
plt.figure(figsize=(15,15))
sns.heatmap(temp_df)
```

```python
# annot : to show the values in the heatmap
temp_df = gap[gap['continent'] == 'Europe'].pivot(index='country',columns='year',values='lifeExp')

plt.figure(figsize=(15,15))
sns.heatmap(temp_df,annot=True,linewidth=0.5, cmap='summer')
```

```python
# Clustermap

# Plot a matrix dataset as a hierarchically-clustered heatmap , it will group the similar rows and columns together
# it will create a dendrogram to show the hierarchical clustering
# it is a figure level function, and it has no axes level function

iris = px.data.iris()
iris
```

```python
sns.clustermap(iris.iloc[:,[0,1,2,3]])
```
